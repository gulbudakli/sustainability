```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('sustainable_development_report_2023.csv')

# En yÃ¼ksek ve en dÃ¼ÅŸÃ¼k 5 overall_score'a sahip Ã¼lkeleri seÃ§in
top_countries = data.nlargest(5, 'overall_score')
bottom_countries = data.nsmallest(5, 'overall_score')

# TÃ¼rkiye'yi iÃ§eren alt kÃ¼me
turkey_and_top_bottom = pd.concat([top_countries, bottom_countries, data[data['country'] == 'TÃ¼rkiye']])

# Yatay barplot oluÅŸturun
plt.figure(figsize=(10, 6))
sns.set_theme(style="whitegrid")
sns.barplot(data=turkey_and_top_bottom, x='overall_score', y='country', palette='pastel')
plt.title('En YÃ¼ksek ve En DÃ¼ÅŸÃ¼k 5 SÃ¼rdÃ¼rÃ¼lebilir KalkÄ±nma Skoruna Sahip Ãœlkeler ve TÃ¼rkiye')
plt.xlabel('SÃ¼rdÃ¼rÃ¼lebilir KalkÄ±nma Skoru')
plt.ylabel('Ãœlkeler')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('sustainable_development_report_2023.csv')

# En yÃ¼ksek ve en dÃ¼ÅŸÃ¼k 5 overall_score'a sahip Ã¼lkeleri seÃ§in
top_countries = data.nlargest(5, 'overall_score')
bottom_countries = data.nsmallest(5, 'overall_score')

# Verileri overall_score'a gÃ¶re sÄ±ralayÄ±n
sorted_data = data.sort_values(by='overall_score', ascending=False)

# TÃ¼rkiye'nin indeksini alÄ±n
turkey_index = sorted_data[sorted_data['country'] == 'TÃ¼rkiye'].index[0]

# TÃ¼rkiye'yi en yÃ¼ksek ve en dÃ¼ÅŸÃ¼k 5 Ã¼lkenin ortasÄ±na yerleÅŸtirin
top_bottom_middle_countries = pd.concat([sorted_data.iloc[:5], sorted_data.iloc[turkey_index:turkey_index+1], sorted_data.iloc[-5:]])

# Yatay barplot oluÅŸturun
plt.figure(figsize=(10, 6))
sns.set_theme(style="whitegrid")
sns.barplot(data=top_bottom_middle_countries, x='overall_score', y='country', palette='pastel')
plt.title('En YÃ¼ksek ve En DÃ¼ÅŸÃ¼k 5 SÃ¼rdÃ¼rÃ¼lebilir KalkÄ±nma Skoruna Sahip Ãœlkeler ve TÃ¼rkiye')
plt.xlabel('SÃ¼rdÃ¼rÃ¼lebilir KalkÄ±nma Skoru')
plt.ylabel('Ãœlkeler')
plt.show()
```
